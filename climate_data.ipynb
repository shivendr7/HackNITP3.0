```python
df_w=pd.read_csv('wind_sp.txt', sep=" ",header=None)
df_w.columns=('jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
df_w;
```

```python
hum=pd.read_csv('humid.txt', sep=" ",header=None)
hum.columns=('jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
hum;
```

```python
cloud=pd.read_csv('cloud.txt', sep=" ",header=None)
cloud.columns=('jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
cloud;
```

```python
rain_mi=pd.read_csv('rain_min.txt', sep=" ",header=None)
rain_mi.columns=('jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
rain_mi;
```

```python
rainy_day=pd.read_csv('rainy_day.txt', sep=" ",header=None)
rainy_day.columns=('jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec")
rainy_day;
```

```python
df=df[['jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec"]]
df;
```

```python
df_min=df_min[['jan','f','ma',"ap","ma","jun","jul","aug","sep","oct","nov","dec"]]
df_min;
```

```python
column_names = ["max_temp(c)", "min_temp(c)", "mean_windspeed(km/hr)","re_hum(%)","cloud(okta)",
                "Mean_rain(mm)","avg._no._rainyday","district"]
z = pd.DataFrame(columns = column_names)
z["max_temp(c)"]=df['jan']
z["min_temp(c)"]=df_min['jan']
z["mean_windspeed(km/hr)"]=df_w['jan']
z["re_hum(%)"]=hum['jan']
z["cloud(okta)"]=cloud["jan"]
z["district"]=["Bhagalpur","chapra","Darbhanga","Dehri","Forbesganj","Gaya","jamui","Motihari","Muzaffarpur","Patna","Purnea","Raxaul","Sabour","Supaul"]
z["Mean_rain(mm)"]=rain_mi['jan'];
z["avg._no._rainyday"]=rainy_day["jan"]
z
```

```python
##   Day wise Data

import random
import numpy as np
lt=[]   #district wise temp
d={} #dict
zt=[]  #temperature as of whole district
for i in zip(z["max_temp(c)"],z["min_temp(c)"],z["district"]):
     lt=[]
     for j in range(1,32):
            if(j<16):
              lt.append(random.uniform(i[1],(i[0]+i[1])/2))
              zt.append(random.uniform(i[1],(i[0]+i[1])/2))
            else:
              lt.append(random.uniform((i[0]+i[1])/2,i[0]))
              zt.append(random.uniform((i[0]+i[1])/2,i[0]))

     arr=np.array(lt)          
     d[i[2]]=lt         
# print(d['Patna']);
```

```python
wind=[] # wind data list
humid=[]  #humid list
cl=[]   #cloud list

for i in zip(z["mean_windspeed(km/hr)"],z["re_hum(%)"],z["cloud(okta)"],z["district"]):
       for j in range(1,32):
            w=random.uniform(-1,1)
            h=random.uniform(-2,2)
            c=random.uniform(-.2,.2)
            wind.append(i[0]+w)
            humid.append(i[1]+h)
            cl.append(i[2]+c)
```

```python
rain=[]  #rain list
day=[]    #avg rainy day list
for i in zip(z["Mean_rain(mm)"],z["avg._no._rainyday"],z["district"]):
    for j in range(1,32):
        h=random.uniform(-2,2)
        rain.append(i[0]+h)
        day.append(i[1])
```

```python
column_names = ["temp(cel)", "mean_windspeed(km/hr)","re_hum(%)","cloud(okta)",
                "rain(mm)","avg._no._rainyday","district"]
raw = pd.DataFrame(columns = column_names)
ld=[] #list of District

#for making inserting district name multiple times
for i,j in d.items():
    for j in range(31):
        ld.append(i)
arr=np.array(ld)

raw['district']=arr
raw["temp(cel)"]=zt
raw["mean_windspeed(km/hr)"]=wind;
raw['re_hum(%)']=humid
raw['cloud(okta)']=cl
raw['rain(mm)']=rain
raw['avg._no._rainyday']=day
raw
```
